```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions
```

```python
driver = webdriver.Chrome(executable_path=r'D:\Selenium_Beautifulsoup\chromedriver.exe')
driver.get('https://inta.tax.gov.ir/Pages/Action/LawsIndex/1')
```

```python
# select bab4
bab = driver.find_element(By.CSS_SELECTOR,'#LabelMainContent > table:nth-child(4) > tbody > tr > td:nth-child(2) > a')
#to get the next page >>> Fasl ha
bab.send_keys(Keys.ENTER) 
```

#  get data from all parts  >>> 4-1 to 4-9

```python
fasl_ha = driver.find_elements(By.CSS_SELECTOR,'#LabelMainContent > table > tbody > tr > td:nth-child(2) > a') 
shomare_fasl_ha = [i for i in range(1,(len(fasl_ha)+1))]





#gereftan etelaat marbot be har fasl
for i in range(0,len(shomare_fasl_ha)):
     
    fasl_ha[i].click() 
    madde_ha = driver.find_elements(By.CSS_SELECTOR,'#LabelMainContent > table > tbody')
    link_madde_ha = [tag.find_element(By.XPATH,'//*[@id="LabelMainContent"]/table/tbody/tr[2]/td[2]/a') for tag in madde_ha]
    shomare_madde_ha = [tag.find_element(By.CSS_SELECTOR,'#LabelMainContent > table > tbody > tr> td > a').get_attribute("text") for tag in madde_ha]
    #gereftan data marbot be har madde az yek fasl
    for j in range(len(link_madde_ha)):
        
        
        # database values
        title_eblaghiye = []
        number_eblaghiye = []
        date_eblaghiye = []
        type_eblaghiye = []
        madde_number = []
        fasl_number = []
        matn_eblaghiye = []
        peyvaste_eblaghiye = []
        mavad_ghanoni_vabaste = []
        
        link_madde_ha[j].click()
        matn_madde = driver.find_elements(By.CSS_SELECTOR,'#LabelMainContent > p')
        matn_madde = [matn.get_attribute("textContent") for matn in matn_madde]
        matn_madde = ''.join(matn_madde)
        
        # ezafe kardan etelaat marbot be madde har fasl
        type_eblaghiye.append('ماده')
        madde_number.append(shomare_madde_ha[j])
        matn_eblaghiye.append(matn_madde )
        title_eblaghiye.append(None)
        number_eblaghiye.append(None)
        fasl_number.append(None)
        date_eblaghiye.append(None)
        peyvaste_eblaghiye.append(None)
        mavad_ghanoni_vabaste.append(None)
        
        filename = f"Bab4-Fasl{i+1}-madde{shomare_madde_ha[j]}.csv"
        
        # peyda kardan eblaghiye hay marbot be har madde az yek fasl
        noe_eblaghiye = driver.find_elements(By.CLASS_NAME,'itemtitle')
        noe_eblaghiye_text = [title.get_attribute("textContent") for title in noe_eblaghiye]
        title_and_link_eblaghiye = dict.fromkeys(noe_eblaghiye_text)
        for k in range(len(noe_eblaghiye_text)):
            key_txt = noe_eblaghiye_text[k]
            if k==len(noe_eblaghiye_text)-1:
                title_and_link_eblaghiye[key_txt] = driver.find_elements(
                    By.XPATH,
                    f'//*[@id="LabelMainContent"]/table[./preceding-sibling::div/a/b[text()="{noe_eblaghiye_text[k]}"]]/tbody/tr/td[2]/a')
            else:
                title_and_link_eblaghiye[key_txt] = driver.find_elements(
                    By.XPATH,
                    f'//*[@id="LabelMainContent"]/table[./preceding-sibling::div/a/b[text()="{noe_eblaghiye_text[k]}"]][./following-sibling::div/a/b[text()="{noe_eblaghiye_text[k+1]}"]]/tbody/tr/td[2]/a')     
         
    
        #gereftan data marbot be har madde az yek fasl
        for m in title_and_link_eblaghiye:
            for n in range(len(title_and_link_eblaghiye[m])):
                title = ' '.join(title_and_link_eblaghiye[m][n].get_attribute("text").split(' ')[1:])
                title_eblaghiye.append(title)
                date = title_and_link_eblaghiye[m][n].get_attribute("text").split(' ')[0][1:11]
                date_eblaghiye.append(date)
                end = len(title_and_link_eblaghiye[m][n].get_attribute("text").split(' ')[0])
                type_ = m
                type_eblaghiye.append(type_)
                madde_number.append(shomare_madde_ha[j])
                fasl_number.append(shomare_fasl_ha[i])
                number = title_and_link_eblaghiye[m][n].get_attribute("text").split(' ')[0][13:end-1]
                number_eblaghiye.append(number)
                
                
                link = title_and_link_eblaghiye[m][n]
                link.click() 
                
                try:
                    peyvast = driver.find_element(By.XPATH,"//*[contains(text(),'پیوست:')]").get_attribute("textContent").split(':')[-1] 
                    peyvaste_eblaghiye.append(peyvast)
                except NoSuchElementException:
                    peyvast=None
                    peyvaste_eblaghiye.append(peyvast)

                    
                    
                
                parentElement=driver.find_element(By.XPATH,'//*[@id="LabelMainContent"]')
                matn_peyvast = parentElement.find_elements(By.XPATH,"*[./following-sibling::div[text()='مواد قانونی وابسته']]")
                matn_peyvast = [matn.get_attribute("textContent") for matn in matn_peyvast]
                matn_peyvast = ''.join(matn_peyvast)
                matn_eblaghiye.append(matn_peyvast)
                
                mavad_vabaste = driver.find_elements(By.XPATH,'//*[@id="LabelMainContent"]/table/tbody/tr/td[2]/a')
                mavad_vabaste = [el.get_attribute('text') for el in mavad_vabaste]
                for made in range(len(mavad_vabaste)):
                    if '-' in mavad_vabaste[made] :
                        mavad_vabaste[made] = mavad_vabaste[made].split('-')[0]
                        
                mavad_ghanoni_vabaste.append(','.join(mavad_vabaste))
                
                
                driver.back()  # bargasht be safheye madde baray entekhab eblaghiye baddi
                
               
                ignored_exceptions=(NoSuchElementException,StaleElementReferenceException,)
                
            
                
                for k in range(len(noe_eblaghiye_text)):
                    key_txt = noe_eblaghiye_text[k]
                    if k==len(noe_eblaghiye_text)-1:
                        title_and_link_eblaghiye[key_txt] =WebDriverWait(
                            driver,1,ignored_exceptions=ignored_exceptions).until(expected_conditions.presence_of_all_elements_located((
                            By.XPATH,
                            f'//*[@id="LabelMainContent"]/table[./preceding-sibling::div/a/b[text()="{noe_eblaghiye_text[k]}"]]/tbody/tr/td[2]/a')))
                
                    else:
                        title_and_link_eblaghiye[key_txt] =WebDriverWait(
                            driver,1,ignored_exceptions=ignored_exceptions).until(expected_conditions.presence_of_all_elements_located((
                            By.XPATH,
                            f'//*[@id="LabelMainContent"]/table[./preceding-sibling::div/a/b[text()="{noe_eblaghiye_text[k]}"]][./following-sibling::div/a/b[text()="{noe_eblaghiye_text[k+1]}"]]/tbody/tr/td[2]/a')))
                
        
        data = pd.DataFrame({
            'ماده' :  madde_number,
            'فصل' : fasl_number,
            'نوع ابلاغیه':type_eblaghiye,
            'موضوع': title_eblaghiye,
            'شماره':number_eblaghiye,
            'پیوست':peyvaste_eblaghiye,
            'تاریخ':date_eblaghiye,
            'متن':matn_eblaghiye,
            'مواد قانونی وابسته':mavad_ghanoni_vabaste,
        })
        data.to_csv(f'D:\\Selenium_Beautifulsoup\\Maliyat_Bab4\\data\\{filename}',index=False,encoding='utf-8-sig')    
        driver.back() # baray entekhab madde baddi az yek fasl be safehey madde ha barmigardad
        ignored_exceptions=(NoSuchElementException,StaleElementReferenceException,)
        link_madde_ha = WebDriverWait(driver,1,ignored_exceptions=ignored_exceptions)\
                        .until(expected_conditions.presence_of_all_elements_located((By.XPATH,'//*[@id="LabelMainContent"]/table/tbody/tr[2]/td[2]/a')))
       
    
    driver.back() # baray raftan be fasl baddi
    ignored_exceptions=(NoSuchElementException,StaleElementReferenceException,)
    fasl_ha = WebDriverWait(driver,1,ignored_exceptions=ignored_exceptions)\
                        .until(expected_conditions.presence_of_all_elements_located((By.CSS_SELECTOR,'#LabelMainContent > table > tbody > tr > td:nth-child(2) > a')))
        
```
